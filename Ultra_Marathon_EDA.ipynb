## Working on Filtered data (df_v2)

```python
# remove (USA) from event name
df_v2['Event name'].str.split('(').str.get(0)
```

```python
df_v2['Event name'] = df_v2['Event name'].str.split('(').str.get(0)
```

```python
df_v2.head()
```

## CLean up Athlete Age

```python
df_v2['athlete_age'] = 2020 - df_v2['Athlete year of birth']
df_v2
```

```python
# remove 'h' from athlete performance
```

```python
df_v2['Athlete performance'] = df_v2['Athlete performance'].str.split(' ').str.get(0)
df_v2
```

```python
# drop columns : Athlete club, Athlete Country, Athlete year of birth, Athelete age category
df_v2 = df_v2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis =1)
df_v2
```

```python
# clean up null values
```

```python
df_v2.isna().sum()
```

```python
df_v2[df_v2['athlete_age'].isna()==1]
```

```python
# drop these na
df_v2 = df_v2.dropna()
df_v2
```

```python
df_v2.isna().sum()
```

```python
# check for duplicates
df_v2[df_v2.duplicated()== True]
df_v2
```

```python
# reset index 
df_v2 = df_v2.reset_index()
df_v2
```

```python
# fix types
df_v2.info()
```

```python
# lets fix athlete_age
df_v2['athlete_age'] = df_v2['athlete_age'].astype(int)
```

```python
df_v2.info()
```

```python
df_v2['Athlete average speed'] = df_v2['Athlete average speed'].astype(float)
```

```python
df_v2.dtypes
```

```python
# rename columns 
df_v2  = df_v2.rename(columns={
    'Year of event':'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length':'race_length',
    'Event number of finishers':'race_number_of_finishers',
    'Athlete performance' : 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed':'athlete_average_speed',
    'Athlete ID': 'athlete_id'
    
})
```

```python
df_v2
```

```python
df_v2.info()
```

```python
# reorder columns
df_v3 = df_v2[['race_day','race_name','race_length','race_number_of_finishers','athlete_id','athlete_gender','athlete_age','athlete_performance','athlete_average_speed']]
```

## Questions I want to find out from the data


```python
# Difference in speed 50k, 50mi male to female
df_v3.groupby(['race_length','athlete_gender'])['athlete_average_speed'].mean()
```

```python
# what age groups are the best in the 50mi race (20 + races min)
df_v3.query('race_length=="50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19')
```

```python
# what age groups are the worst in the 50mi race (20 + races min)
df_v3.query('race_length=="50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>19')
```

```python
# what age groups are the best in the 50km race (20 + races min)
df_v3.query('race_length=="50km"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = False).query('count>19')
```

```python
# what age groups are the worst in the 50km race (20 + races min)
df_v3.query('race_length=="50km"').groupby('athlete_age')['athlete_average_speed'].agg(['mean','count']).sort_values('mean', ascending = True).query('count>19')
```

```python
# Seasons for the data - Slower in summer than winter?
#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2
```

```python
df_v3['race_month'] = df_v3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df_v3.head()
```

```python
df_v3['race_season'] = df_v3['race_month'].apply(lambda x: 'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter')
```

```python
df_v3.head(10)
```

```python
values = df_v3['race_season'].value_counts()
print(values)
```

```python
df_v3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending= False)
```
